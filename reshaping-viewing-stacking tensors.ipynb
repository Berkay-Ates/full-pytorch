## Reshaping, Stacking, Squezing and Unsquezing TENSOR

* **Reshaping** -> reshapes an input tensor to a defined shape
* **View** -> return a view of an input tensor of certain shape but keep the same memory as the original tensor 
* **Stacking** -> combine multiple tensors on top of each other (vstack) or side by side (hstack)
* **Squeeze** -> removes all '1' dimensions from tensor 
* **UnSqueeze** -> add '1' dimension to a target tensor 
* **Permute** -> Return a view of the input with dimensions permuted (swapped) in a certain way


```python
x = torch.arange(2,90,12)
```

```python
x, x.shape
```

### Add an extra dimension

- New shape have to be compatible with the previous shape

```python
x_reshaped = x.reshape((1,8))
```

### View 

- Changing z changes x (because a view of a tensor shares the same memory as the original input)

```python
z = x.view(1,8)
z,x
```

```python
z[:,0] = 5
```

```python
z,x
```

### Permute

```python
x_original = torch.rand(size=(224,224,3)) #* [height,weight,color Channel]
```

```python
plt.imshow(x_original)
```

```python
x_permuted = x_original.permute(2,0,1)
```

```python
print(f'previous shape: {x_original.shape}')
print(f'permuted shape: {x_permuted.shape}')
```

### **:** (semi colon) can be used to select all the target dimension

```python
x[:,0]
```

```python
# Get all values of 0th and 1st dimension but only 1 of 2nd dimension
x[:,:,2]
```

```python
# Get all values of the 0 dim but only the 1 index value of the 1st and 2nd dimension 
x[:,1,1]
```

```python
# Get index 0 of 0th and 1th dimension and all values of 2nd dimension 
x[0,0,:]
```
